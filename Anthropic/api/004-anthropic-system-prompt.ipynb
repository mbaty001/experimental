```python
from anthropic import Anthropic
client = Anthropic()
model = "claude-haiku-4-5"

messages = []

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages, system_prompt=None):

    params = {
        "model": model,
        "max_tokens": 10,
        "messages": messages,
    }
    if system_prompt:
        params["system"] = system_prompt

    message = client.messages.create(
        **params
    )
    return message.content[0].text
```

```python

# Start with an empty message list
messages = []

# Add collected prompt to the conversation history
add_user_message(messages, "Solve the equation x^2 - 5x + 6 = 0")

# Get Claude's response
answer = chat(messages)
print("Claude's response:", answer)

# Add system prompt
system_prompt = "You are a math tutor. Provide a step-by-step solution to the equation."
answer_with_system = chat(messages, system_prompt=system_prompt)
print("Claude's response with system prompt:", answer_with_system)
```

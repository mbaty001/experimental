````python
from anthropic import Anthropic
client = Anthropic()
model = "claude-haiku-4-5"

messages = []

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages, system_prompt=None):

    params = {
        "model": model,
        "max_tokens": 10,
        "messages": messages,
    }
    if system_prompt:
        params["system"] = system_prompt

    message = client.messages.create(
        **params
    )
    return message.content[0].text
````

````python

# Start with an empty message list
messages = []

prompt = """Generate three different sample of AWS cli command. Each should be very short. Do not comment the commands. Do not add any text other than the commands. Each command should be on a separate line."""
add_user_message(messages, prompt)
add_assistant_message(messages, "```bash")

response= client.messages.create(
    model=model,
    messages=messages,
    max_tokens=1000,
    stop_sequences=["```"]
)

print(response.content[0].text)
````

````python

# Start with an empty message list
messages = []

add_user_message(messages, "Generate a 1 sentence description of a cat in the style of a haiku.")

with client.messages.stream(
    model=model,
    messages=messages,
    max_tokens=10,
) as stream:

    for text in stream.text_stream:
        print (text)

    print("whole response:", stream.get_final_message())
````
